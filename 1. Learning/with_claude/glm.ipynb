```python
# Dataset: Titanic (built into seaborn)
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import glm
import numpy as np
import pandas as pd

# Load data
titanic = sns.load_dataset('titanic')

# Task: Predict survival using logistic regression
# Features: age, fare, pclass, sex

# Check the features
titanic.head()
```

```python

# TODO 1: Fit logistic regression
model = glm('survived ~ age + fare + pclass + sex', data=titanic, family=sm.families.Binomial()).fit()
print(model.summary())

# TODO 2: Interpret coefficients
coefficients = model.params
print("Coefficients:\n", coefficients)

# TODO 3: What's the odds ratio for being female vs male?
odds_ratio_female_vs_male = 1 / np.exp(coefficients['sex[T.male]'])
print("Odds ratio for being female vs male:", odds_ratio_female_vs_male)

# TODO 4: Predict probability for: male, age=30, pclass=3, fare=10
# Note that in statsmodels, .predict() returns probabilities by default for logistic regression
new_data = pd.DataFrame({'age': [30], 'fare': [10], 'pclass': [3], 'sex': ['male']})
probabilities = model.predict(new_data)
print("Predicted probability of survival for male, age=30, pclass=3, fare=10:", probabilities[0])
```
